####class vs. object vs. instance

  - Class: A class is an object creation blueprint or template. It specifies the characteristics and actions (methods) that an object of that class will possess.
  - Object: A class instance is what's known as an object. It stands for a particular embodiment of the class, complete with special characteristics and behavior.
  - Instance:  an object is a single occurrence of that item. When a class is instantiated, a unique realization of the class is produced.


```python
#Class
class cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def meow(self):
        return "meow!"
```

```python
#Object and Instance
a_cat = cat("Buddy", 3)
```

##method vs. function

Methods are a basic idea in object-oriented programming that are used to encapsulate activity within classes and allow objects to carry out particular tasks. However, functions are stand-alone pieces of code that can be utilized inside a program's larger framework.


```python
#calling method
message = a_cat.meow()
print(message)
```

```python
#function
def sound():
  return "Aho!"

print(sound())
```

##attribute vs. variable

  - Variable: Named storage location for holding data values within a program.
  - Attribute: Data associated with an object that defines its state and  
    properties within a class.

```python
#variables
name = "John"
age = 20
```

```python
#Attributes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person1 = Person("Alice", 30)

print(person1.name) #Accessing Attributes
print(person1.age) #Accessing Attributes
```

##Explain the difference in focus and approach of procedural and OOP paradigms

### Programming in Procedures:
- focuses on processes or operations involving data.
- emphasizes using functions to carry out tasks step-by-step.
- breaks down issues into their component parts using a top-down method.
- Functions and data are two different things.
###Programming that is object-oriented (OOP):
- focuses on representing things from the real world as objects with properties
and functions.
- highlights the concept of encapsulation, which is the grouping of functions and data into objects.
- uses classes to build object blueprints, enabling inheritance-based code reuse.
- helps to advance modular design and abstraction by presenting ideas as objects.


##Explain the advantage of using a class over using modules which only contain a collection of data and functions

Python classes, especially when used in an object-oriented paradigm, are preferable to modules in terms of code organization, reusability, and maintainability. Classes allow abstraction and modularity, support inheritance and polymorphism, encapsulate data and behavior into coherent units, and follow important design guidelines. You may create scalable, reliable programs that are easier to comprehend, expand, and manage over time by making good use of classes. Using classes in object-oriented programming is a potent paradigm that enables programmers to solve challenging issues elegantly and effectively.
