Write a program to read through a file and output the number of lines in a file.

```python
files = open('John.txt', 'r')
print(files)
count = 0
for line in files:
  line = line.rstrip()
  count = count + 1
print("Output the number of lines in a file: ", count)
```

## CHALLENGE

from [PEW] E8-2

Write a program to prompt for a file name, and then read through the file and look for lines of the form:

When you encounter a line that starts with "X-DSPAM-Confidence:" pull apart the line to extract the floating-point number on the line. Count these lines and then compute the total of the spam confidence values from these lines. When you reach the end of the file, print out the average spam confidence.

```python
filename = input("Enter a filename: ")
count = 0
total = 0
while(filename != ""):
  try:
    f = open(filename, 'r')
    for line in f:
      line = line.strip()
      if line.startswith('X-DSPAM-Confidence:'):
        colon_index = line.find(':')
        if colon_index != -1:
          num_str = line[colon_index + 1:].strip()
          if num_str:
            confidence_value = float(num_str)
            total += confidence_value
            count += 1
    if count > 0:
        average = total / count
        print(f"The average of spam confidence is {average:.4f}.")
        break
    else:
        print(f"X-DSPAM-Confidence does not exist in the content of found in the '{filename}'.")
        break
  except FileNotFoundError:
      print(f"No file exists: '{filename}'")
      continue
  except Exception as e:
      print(f"There is an error {e}")
      continue
```
